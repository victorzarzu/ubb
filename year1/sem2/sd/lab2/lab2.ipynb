```python
x = var('x')
y = function('y')(x)
yy = var('yy')
k = var('k')
a,b = var('a,b')
#_C = var('_C')
```

```python
#1
```

```python
#a

eqd = diff(y, x) == 2*x(1 + y^2)
sold = desolve(eqd, y)
show(sold)
sol = solve(sold, y)
show(sol)

f(x, _C) = sol[0].rhs();
show(f)

plot(s(x, 4), -2, 2, detect_poles = 'True', ymin = -3, ymax = 3)
```

```python
#b

eqd = (x^2 - 1)*diff(y, x) + 2*x*y^2 == 0
sold = desolve(eqd, y, show_method = True)
show(sold)

sol = solve(sold[0], y)
show(sol)

f(x, _C) = sol[0].rhs();
plot(f(x, 0), 5, 100)
```

```python
#c

eqd = 2*(x^2)*diff(y, x) == x^2 + y^2
sold = desolve(eqd, y)
show(sold)

sol = solve(sold, y)
show(sol)

f1(x, _C) = sol[0].rhs()
f2(x, _C) = sol[1].rhs()

plot([f1(x, 4), f2(x, 4)], 5, 8, color = ['blue', 'red'])
```

```python
#d

eqd = diff(y, x) == -(x + y) / y
sold = desolve(eqd, y)
show(sold)

sol = solve(sold, y)
show(sol)

f(x, yy, _C) = sold.substitute(y == yy)
show(f)

implicit_plot(f(x, yy, 5), (x, -1, 1), (yy, -1, 1))
```

```python
#e

eqd = diff(y, x, 2) + y == sin(x) + cos(x)
sold = desolve(eqd, y)
show(sold)

f(x, _K1, _K2) = sold
plot(f(x, 0, 5), x, -1, 1)
```

```python
#f

eqd = diff(y, x, 2) - y == e^(2*x)
sold = desolve(eqd, y)
show(sold)

f(x, _K1, _K2) = sold
plot(f(x, 0, 5), x, -1, 1)
```

```python
#g

eqd = diff(y, x, 2) - diff(y, x) == 1/(1 + e^x)
sold = desolve(eqd, y)
show(sold)

f(x, _K1, _K2) = sold
plot([f(x, 1, 2), f(x, 5, 10), f(x, 0, 0)], x, -2, 2, detect_poles = True)
```

```python
#b

eqd = diff(y, x) == (1/(1 - x^2))*y + 1 + x
solc(x) = desolve(eqd, y, ics = [2, 0])
show(solc)

plot(solc(x), x, 2, 100)
```

```python
#d

eqd = diff(y, x, 2) - 5*diff(y, x) + 4*y == 0
solc(x) = desolve(eqd, y, ics = [0, 5, 8])
show(solc(x))

plot(solc(x), x, -0.5, 0.5, detect_poles = True, ymin = -7, ymax = 7)
```

```python
#e

eqd = diff(y, x, 2) - 4*diff(y, x) + 5*y == 2*x^2*e^x
solc(x) = desolve(eqd, y, ics = [0, 2, 3])
show(solc(x))

plot(solc(x), x, -pi/4, pi/4)
```

```python
#f

eqd = diff(y, x, 2) + 4*y == 4*(sin(2*x) + cos(2*x))
solc(x) = desolve(eqd, y, ics = [pi, 2*pi, 2*pi])
show(solc(x))

plot(solc(x), x, -4, 4)
```

```python
#b

f(x, k, _C) = solc
plot([f(x, 1, 0), f(x, 1, 1), f(x, 1, 2)], x, -1, 1, detect_poles = True, ymin = -3, ymax = 3)
```

```python
#4

eqd = diff(y, x) == y + a*x + b
solc(x) = desolve(eqd, [y, x], ics = [0, 1])
show(solc)

sol = solve([solc(1) == -5 + 4*e, solc(2) == -7 + 4*e^2], a, b)
show(sol)

f = solc.substitute(a == 2, b == 1)
plot(f(x), x, -5, 5)
```
